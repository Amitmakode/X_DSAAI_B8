## Reduce

```python
from functools import reduce
```

```python
# Example - find product of all numbers
numbers = [2, 3, 4]
```

```python
# Using lambda with reduce
product = reduce(lambda x, y: x * y, numbers)
```

```python
product
```

```python
# Sample data
original_list = [1, 2, 3, 4, 5, 10, 15, 20]
```

```python
# 1. Convert to cubes using map()
cubes = list(map(lambda x: x**3, original_list))
print(f"Cubes: {cubes}")
```

```python
# 2. Filter numbers divisible by 5
divisible_by_5 = list(filter(lambda x: x % 5 == 0, original_list))
print(f"Divisible by 5: {divisible_by_5}")
```

```python
# 3. Find product using reduce()
product = reduce(lambda x, y: x * y, [1, 2, 3, 4, 5])
print(f"Product of 1,2,3,4,5: {product}")
```

```python
try:
    x = int(input("Enter a number: "))
    y = 10 / x
    print(f"Result: {y}")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
except ZeroDivisionError:
    print("Cannot divide by zero. Please enter a non-zero number.")
except Exception as e:
    print(f"Error occurred: {e}")
```

```python
class InsufficientBalanceError(Exception):
    pass

balance = 5000
withdraw = 7000

try:
    if withdraw > balance:
        raise InsufficientBalanceError("Not enough balance")
except InsufficientBalanceError as e:
    print(e)
```

```python
file_names = ["file1.txt", "file2.txt", "file3.txt"]

for file_name in file_names:
    try:
        file = open(file_name, "r")
        content = file.read()
        print(f"Content of {file_name}: {content}")
    except FileNotFoundError:
        print(f"File '{file_name}' not found.")
    except IOError as e:
        print(f"Error reading file '{file_name}': {e}")
    finally:
        if 'file' in locals():
            file.close()
```
