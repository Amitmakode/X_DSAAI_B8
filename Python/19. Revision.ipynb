```python
class Employee:
    company = "TechCorp"  # Class variable
    employee_count = 0    # Class variable
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.employee_count += 1
    
    @classmethod
    def get_company(cls):  # Class method
        """Returns company name"""
        return cls.company
    
    @classmethod
    def get_employee_count(cls):  # Class method
        """Returns total employee count"""
        return cls.employee_count
    
 
    def create_intern(cls, name):  # Alternative constructor
        """Creates an intern with default salary"""
        return cls(name, 20000)
```

```python
Employee("abc", 10000)
```

```python
class MathUtils:
    
    @staticmethod
    def add(a, b):  # Static method
        """Adds two numbers"""
        return a + b
    
    @staticmethod
    def is_even(num):  # Static method
        """Checks if number is even"""
        return num % 2 == 0
    
    @staticmethod
    def factorial(n):  # Static method
        """Calculates factorial"""
        if n <= 1:
            return 1
        return n * MathUtils.factorial(n - 1)
```

```python
MathUtils.is_even(5)
```
