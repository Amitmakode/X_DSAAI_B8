```python
# Parent class
class Vehicle:
    def general_info(self):
        print("Vehicles are used for transportation.")

# Child class 1 (inherits from Vehicle)
class Car(Vehicle):
    def car_info(self):
        print("Car has 4 wheels.")

# Child class 2 (inherits from Vehicle)
class Bike(Vehicle):
    def bike_info(self):
        print("Bike has 2 wheels.")

# Grandchild class (inherits from both Car and Bike)
class Electric(Bike, Car ):
    def electric_info(self):
        print("Electric bike runs on battery.")

```

```python
Electric.mro()
```

## Encapsulation

```python
class Car:
    def __init__(self, brand, speed=0):
        self.__brand = brand  # Private attribute
        self.__speed = speed  # Private attribute

    def accelerate(self, increase):
        self.__speed += increase

    def get_speed(self):
        return self.__speed

    def get_brand(self):
        return self.__brand
```

```python
obj = Car("BMW")
```

```python
obj.get_brand()
```

```python
obj.get_speed()
```

```python
obj.accelerate(40)
```

```python
obj.get_speed()
```

## Polymorphism  

```python
class Vehicle:
    def move(self):
        return "Vehicle is moving."

class Car(Vehicle):
    def move(self):  # Overrides parent method
        return "Car is driving on the road."

class Boat(Vehicle):
    def move(self):  # Overrides parent method
        return "Boat is sailing on water."
```

```python
# Usage
vehicles = [Vehicle(), Car(), Boat()]

for i in vehicles:
    print(i.move())
```

## super()  key word 

```python
# Parent class
class Vehicle:
    def general_info(self):
        print("Vehicles are used for transportation.")

# Child class 1  
class Car(Vehicle):
    def general_info(self):
        super().general_info()
        print("car class--- Vehicles are used for transportation.")
    
    def car_info(self):
        print("Car has 4 wheels.")
```

```python
obj3 = Vehicle()
```

```python
obj3.general_info()
```

```python
obj4 = Car()
```

```python
obj4.general_info()
```
