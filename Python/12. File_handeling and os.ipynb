```python
i = 1
while i <= 10:
    file_name = f'file{i}.txt'
    if os.path.exists(file_name):
        os.remove(file_name)
        print(f'Deleted {file_name}')
    else:
        print(f'{file_name} does not exist')
    i += 1
```
